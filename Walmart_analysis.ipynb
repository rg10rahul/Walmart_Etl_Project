```python
#extract csv from zip file
import zipfile
file_ref = zipfile.ZipFile("../walmart-10k-sales-datasets.zip")
file_ref.extractall()
file_ref.close()
```

```python
#read data from csv file
import pandas as pd
df = pd.read_csv("Walmart.csv")
df.head(10)
```

```python
#find all duplicates
df.duplicated().sum()
```

```python
#drop duplicates
df.drop_duplicates(inplace=True)
#verify
df.duplicated().sum()
```

```python
#remove all missing values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head(10)
```

```python
#convert unit price column datatype to float
df['unit_price'] = df['unit_price'].str.replace("$"," ").astype(float)
#verfiy
df.info()
```

```python
df.head()
```

```python
#adding total sales column
df['total'] = df['unit_price']*df['quantity']
#verify
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
#converting column names to lowercase
df.columns = df.columns.str.lower()
#verify
df.columns
```

```python
df.info()
```

```python
#saving transformed csv file
df.to_csv("Walmart_clean_data.csv",index=False)
```
